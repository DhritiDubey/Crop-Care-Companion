```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
pca2 = pd.read_csv('Data_PCA_N=2.csv')
pca3 = pd.read_csv('Data_PCA_N=3.csv')
pca4 = pd.read_csv('Data_PCA_N=4.csv')
```

```python
from sklearn.ensemble import RandomForestClassifier
rfc = RandomForestClassifier(class_weight='balanced_subsample', n_estimators=400)
```

```python
pca2.columns
```

```python
from sklearn.model_selection import train_test_split
Xtrain, Xtest, Ytrain, Ytest = train_test_split(pca2[pca2.columns[1:-1]],pca2[pca2.columns[-1]],test_size=0.3,random_state=14)
Xtrain, Xval, Ytrain, Yval = train_test_split(Xtrain, Ytrain, test_size=0.25,random_state=14)
```

```python
from sklearn.metrics import f1_score, accuracy_score, confusion_matrix
```

```python
rfc = rfc.fit(Xtrain,Ytrain)
Ypred = rfc.predict(Xtest)
plt.figure(figsize=(20,10))
sns.heatmap(confusion_matrix(Ypred,Ytest),annot=True)
plt.show()
test2_acc_score = []
test2_f1score = []
val2_acc_score=[]
val2_f1score=[]
print('Test - Accuracy: '+str(accuracy_score(Ypred,Ytest)))
print("Test - F1-Score: "+str(f1_score(Ypred,Ytest,average='weighted')))
print("Validation - Accuracy: "+str(accuracy_score(rfc.predict(Xval),Yval)))
print("Validation - F1-Score: "+str(f1_score(rfc.predict(Xval),Yval,average='weighted')))
test2_acc_score.append(accuracy_score(Ypred,Ytest))
test2_f1score.append(f1_score(Ypred,Ytest,average='weighted'))
val2_acc_score.append(accuracy_score(rfc.predict(Xval),Yval))
val2_f1score.append(f1_score(rfc.predict(Xval),Yval,average='weighted'))
```

```python
from sklearn.model_selection import train_test_split
Xtrain, Xtest, Ytrain, Ytest = train_test_split(pca3[pca3.columns[1:-1]],pca3[pca3.columns[-1]],test_size=0.3,random_state=14)
Xtrain, Xval, Ytrain, Yval = train_test_split(Xtrain, Ytrain, test_size=0.25,random_state=14)
rfc = rfc.fit(Xtrain,Ytrain)
Ypred = rfc.predict(Xtest)
plt.figure(figsize=(20,10))
sns.heatmap(confusion_matrix(Ypred,Ytest),annot=True)
plt.show()
test3_acc_score = []
test3_f1score = []
val3_acc_score=[]
val3_f1score=[]
print('Test - Accuracy: '+str(accuracy_score(Ypred,Ytest)))
print("Test - F1-Score: "+str(f1_score(Ypred,Ytest,average='weighted')))
print("Validation - Accuracy: "+str(accuracy_score(rfc.predict(Xval),Yval)))
print("Validation - F1-Score: "+str(f1_score(rfc.predict(Xval),Yval,average='weighted')))
test3_acc_score.append(accuracy_score(Ypred,Ytest))
test3_f1score.append(f1_score(Ypred,Ytest,average='weighted'))
val3_acc_score.append(accuracy_score(rfc.predict(Xval),Yval))
val3_f1score.append(f1_score(rfc.predict(Xval),Yval,average='weighted'))
```
